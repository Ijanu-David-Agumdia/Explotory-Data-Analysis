```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read the data set 
df_data = pd.read_csv('Book.csv')

#Printing the data
df_data
```

Missing Values

```python
# Checking the missing values
df_data.isnull().sum()
```

```python
df_data.drop(['Tax_amount', 'Price_After_Tax', 'Unnamed: 0'], inplace=True, axis=1)
```

```python
df_data
```

```python
df_data.isna().sum()
```

Handling Outliers

Univariate Analysis

```python
df_data.groupby('Category').mean()
```

```python
df_data.groupby('Category').mean().plot()
```

Bivariate Analysis

1. Numeric_Numeric Analysis

Scatter Plot

```python
#Plot the scatter plot
plt.scatter(df_data.Price, df_data.Avilability)
plt.show()

#plot the scatter plot of balance and age variable in data
df_data.plot.scatter(x='Avilability', y='Price')
plt.show()
```

Pair Plot

```python
#Pair Plot
sns.pairplot(data = df_data, vars=['Avilability', 'Price', 'Category'])
plt.show()
```

Correlation Matrix

```python
# Creating a matrix using age, salary, balance as rows and columns
df_data[['Category', 'Avilability', 'Price']].corr()

#Plot the correlation matrix of salary, balance and age in dataframe
sns.heatmap(df_data[['Category', 'Avilability', 'Price']].corr(), annot=True)
plt.show()
```

2. Numeric-Categorical Analysis

```python
#Plot the boxplot of salary and response
sns.boxplot(df_data.Price, df_data.Avilability)
plt.show()
```

Multivariate Analysis

```python
df_result = pd.pivot_table(data=df_data, index='Category', columns='Avilability', values='Price')
print(df_result)
#Creating heatmap of education vs marital vs response rate
sns.heatmap(df_result, annot=True, cmap = 'RdYlGn')
plt.show()
```
